```python
# read the latency of three npz
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = 'cifar-10'
client_lat = 'client_latency_' + dataset +'.npy'
network_lat = 'network_latency_' + dataset +'.npy'
server_lat = 'server_latency_' + dataset +'.npy'

client_lat = np.load(client_lat)
network_lat = np.load(network_lat)
server_lat = np.load(server_lat)

latency = client_lat + network_lat + server_lat

accuracy = 'jpg_acc_' + dataset +'.npy'
accuracy = np.load(accuracy)

o_latency = 'network_latency_bmp_' + dataset +'.npy'
o_latency = np.load(o_latency)

o_latency = o_latency + server_lat[:,-1]

o_accuracy = 'original_acc_' + dataset +'.npy'
o_accuracy = np.load(o_accuracy)
```

```python
import pandas as pd
dict_latency = {'cr':[], 'iname':[], 'latency':[], 'acc':[]}
h, w = latency.shape
for i in range(latency.shape[0]):
    for j in range(latency.shape[1]):
        cr = int(j*5+5)
        iname = int(i)
        lat = latency[i,j]
        acc = accuracy[i,j]
        dict_latency['cr'].append(cr)
        dict_latency['iname'].append(iname)
        dict_latency['latency'].append(lat)
        dict_latency['acc'].append(acc)
    cr = 100
    iname = int(i)
    lat = o_latency[i]
    acc = o_accuracy[i]
    dict_latency['cr'].append(cr)
    dict_latency['iname'].append(iname)
    dict_latency['latency'].append(lat)
    dict_latency['acc'].append(acc[0])

# dict_latency_refine = {'cr':[], 'latency':[], 'acc':[]}
# for cr in range(20):
#     cr_i = cr*5+5
#     dict_latency_refine['cr'].append(cr_i)
#     lat_avg = 0
#     acc_avg = 0
#     count = 0
#     for cr2 in range(len(dict_latency['cr'])):
#         if dict_latency['cr'][cr2] == cr_i:
#             count += 1
#             lat_avg += dict_latency['latency'][cr2]
#             acc_avg += dict_latency['acc'][cr2]
#     lat_avg /= count
#     acc_avg /= count
#     dict_latency_refine['latency'].append(lat_avg)
#     dict_latency_refine['acc'].append(acc_avg)

# df = pd.DataFrame(dict_latency_refine)
df1 = pd.DataFrame(dict_latency)
sns.lineplot(x="cr", y="acc", data=df1)
```
