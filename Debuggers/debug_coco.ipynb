```python
# this is a debugger for coco
import torch
import torchvision
from torchvision import transforms
def Dataloader_coco(r_path, a_path):
    cocodataset = torchvision.datasets.CocoDetection(
        root=r_path, 
        annFile=a_path,
        transform=transforms.ToTensor())
    return cocodataset
ds = Dataloader_coco('/data/anp407/coco/train2017', '/data/anp407/coco/annotations/instances_train2017.json')
```

```python
for i, (img, target) in enumerate(ds):
    print(i)
    print(img)
    print(target)
    if i == 0:
        break
```
